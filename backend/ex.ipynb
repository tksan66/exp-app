```python
import fitz
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    for page in range(len(doc)):
        text = doc[page].get_text()
        print(f"#############{page}##############")
        print(text)
# PDFファイルのパスを指定してテキストを抽出
pdf_path = "./data/統合報告書.pdf"
extract_text_from_pdf(pdf_path)
```
